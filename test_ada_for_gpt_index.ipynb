```python
import os
os.environ["OPENAI_API_KEY"] = "YOUR_OPENAI_API_KEY"
```

```python
from gpt_index import GPTSimpleVectorIndex, SimpleDirectoryReader, PromptHelper

# インデックスの作成
documents = SimpleDirectoryReader('data').load_data()
```

```python
# インデックスの生成
index = GPTSimpleVectorIndex(
    documents=documents,
    prompt_helper=PromptHelper(
        max_input_size=4000,  # LLM入力の最大トークン数
        num_output=256,  # LLM出力のトークン数
        chunk_size_limit=500,  # チャンクのトークン数
        max_chunk_overlap=0,  # チャンクオーバーラップの最大トークン数
        separator="。"  # セパレータ
    ),
    verbose=True
)
```

```python
print(index.query(
    "後藤ひとりはどういう性格？", 
    verbose=True
))
```

```python
print(index.query(
    "ハリーの性格は？", 
    verbose=True
))
```

```python
print(index.query(
    "who is the guitar hero？", 
    verbose=True,
    similarity_top_k=2
))
```
